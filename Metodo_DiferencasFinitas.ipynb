# Método de Diferenças Finitas

## Equação do Calor

```python
def equacao_calor(k, L, tf, nx, nt):
    """ Essa função aplica o método de diferenças finitas para estimar a variação do calor
    em uma barra, ou seja, no caso unidimensional. Nesse caso, temos a aplicação do método
    de diferenças finitas próxima para o tempo e centrada para o espaço.
    
    Args:
    
        k: coeficiente de difusividade térmica.
        L: tamanho da barra ou valor final do intervalo de x.
        tf: valor final de tempo.
        nx: número de pontos verificados da barra.
        nt: número de valores de tempo verificados.
        
    Return:
        
        u: matriz com os valores da função de calor u.
    """
    
    dt = tf / (nt - 1) # Intervalo de tempo da verificação
    dx = L / (nx - 1) # Intervalo de espaço na barra
    constante = k * dt / dx**2 # Constante que resume a relação encontrada na aplicação do método
    
    u = np.zeros((nt, nx)) # Definindo a matriz u que recebem os pontos
    x = np.linspace(0, L, nx) # Definindo os pontos da barra
    
    u[0, :] = x * ((1 - x)**2) # Aplicando a condição inicial
    
    u[:, 0], u[:,-1] = 0.0, 0.0# Aplicando as condições de contorno
    
    for i in range(1, nt): # Tempo - Excluindo o início
        for j in range(1, nx - 1): # Espaço - Excluindo os contornos
            # Aplicando o método
            u[i, j] = u[i - 1, j] + constante * (u[i - 1, j + 1] - 2 * u[i - 1, j] + u[i - 1, j - 1])

    return u
```

```python
# Parâmetros Iniciais

k = 0.5
L = 1.0
tf = 1.0
nt = 200
nx = 10
```

```python
# Grid para o gráfico

x = np.linspace(0, L, nx)
t = np.linspace(0, tf, nt)

X, T = np.meshgrid(x, t)

# Criar a figura e os subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 10), subplot_kw={'projection': '3d'})

# Plotar no primeiro subplot
k = 0.5
u = equacao_calor(k, L, tf, nx, nt)

surf1 = axs[0].plot_surface(X, T, u, cmap='plasma')
axs[0].set_xlabel('x', labelpad=10)  # Adiciona um espaçamento entre o rótulo e o eixo x
axs[0].xaxis.label.set_rotation(45)  # Rotação da label do eixo x em graus
axs[0].set_ylabel('tempo', labelpad=10)
axs[0].set_zlabel('u(x,t)', labelpad=10)
axs[0].set_title(f'k = {k}')

# Plotar no segundo subplot
k = 0.25
u = equacao_calor(k, L, tf, nx, nt)

surf2 = axs[1].plot_surface(X, T, u, cmap='plasma')
axs[1].set_xlabel('x', labelpad=10)  # Adiciona um espaçamento entre o rótulo e o eixo x
axs[1].xaxis.label.set_rotation(45)  # Rotação da label do eixo x em graus
axs[1].set_ylabel('tempo', labelpad=10)
axs[1].set_zlabel('u(x,t)', labelpad=10)
axs[1].set_title(f'k = {k}')

# Plotar no terceiro subplot
k = 0.1
u = equacao_calor(k, L, tf, nx, nt)

surf2 = axs[2].plot_surface(X, T, u, cmap='plasma')
axs[2].set_xlabel('x', labelpad=10)  # Adiciona um espaçamento entre o rótulo e o eixo x
axs[2].xaxis.label.set_rotation(45)  # Rotação da label do eixo x em graus
axs[2].set_ylabel('tempo', labelpad=10)
axs[2].set_zlabel('u(x,t)', labelpad=10)
axs[2].set_title(f'k = {k}')

# Ajustar os subplots para melhor visualização
plt.subplots_adjust(wspace=0.3)  # Ajustar o espaçamento entre subplots

# Adicionar uma colorbar
cbar = fig.colorbar(surf2, ax=axs, pad=0.05, shrink=0.5, aspect=10)
cbar.ax.set_ylabel('Temperatura', fontsize=15)  # Rótulo da colorbar
cbar.ax.xaxis.set_label_coords(1, 0.)  # Posicionamento do rótulo da colorbar
cbar.ax.yaxis.set_label_coords(0.5, 0.0)  # Posicionamento do rótulo da colorbar
cbar.ax.yaxis.label.set_rotation(0)  # Rotação do rótulo da colorbar
cbar.ax.tick_params(labelsize=13)  # Tamanho dos ticks da colorbar
cbar.ax.yaxis.set_tick_params(width=1)  # Espessura das linhas dos ticks da colorbar
cbar.ax.yaxis.get_offset_text().set_fontsize(13)  # Tamanho do texto do valor máximo/minimo da colorbar
cbar.ax.tick_params(axis='y', length=3, width=0.5)  # Comprimento e espessura dos ticks da colorbar

# Nomeia o gráfico
#plt.title('Mapa de calor da barra ao longo do tempo')

# Exibir o gráfico
plt.show()
```
